```python
import fastf1
```

```python
fastf1.Cache.enable_cache('../backend/src/cache/fastf1')
fastf1.set_log_level('DEBUG')
```

```python
schedule = fastf1.get_event_schedule(2024)
schedule["EventFormat"]
```

```python
race_events = schedule[schedule['EventFormat'].isin(
                ['conventional', 'sprint'])]
race_events
```

```python
last_round_number = race_events['RoundNumber'].max()
last_round_number
```

```python
session = fastf1.get_session(2024, last_round_number, 'R')
session.load()
```
